```python
GK = f20_dream[f20_dream['position'].str.contains('GK')].nlargest(3,'rating')
LB = f20_dream[f20_dream['position'].str.contains('LB')].nlargest(3,'rating')
RB = f20_dream[f20_dream['position'].str.contains('RB')].nlargest(3,'rating')
CB = f20_dream[f20_dream['position'].str.contains('CB')].nlargest(6,'rating') #2x CB
LM = f20_dream[f20_dream['position'].str.contains('LM')].nlargest(3,'rating')
RM = f20_dream[f20_dream['position'].str.contains('RM')].nlargest(3,'rating')
CM = f20_dream[f20_dream['position'].str.contains('CM')].nlargest(3,'rating') 
ST = f20_dream[f20_dream['position'].str.contains('ST')].nlargest(9,'rating') #3x strikers
frames = (GK, LB, RB, CB, LM, RM, CM, ST)
dream_team = pd.concat(frames, axis=0, ignore_index=True)
```

```python
dream_team[['name', 'position', 'preferred_foot', 'rating', 'nationality', 'club', 'value_M', 'wage_K']]
```

```python
GK_on = dream_team[dream_team['position'].str.contains('GK')].nlargest(1,'rating')
LB_on = dream_team[dream_team['position'].str.contains('LB')].nlargest(1,'rating')
RB_on = dream_team[dream_team['position'].str.contains('RB')].nlargest(1,'rating')
CB_on = dream_team[dream_team['position'].str.contains('CB')].nlargest(2,'rating') #2x CB
LM_on = dream_team[dream_team['position'].str.contains('LM')].nlargest(1,'rating')
RM_on = dream_team[dream_team['position'].str.contains('RM')].nlargest(1,'rating')
CM_on = dream_team[dream_team['position'].str.contains('CM')].nlargest(1,'rating') 
ST_on = dream_team[dream_team['position'].str.contains('ST')].nlargest(3,'rating') #3x strikers
```

```python
frames_on = (GK_on, LB_on, RB_on, CB_on, LM_on, RM_on, CM_on, ST_on)
dream_team_on = pd.concat(frames_on, axis=0, ignore_index=True)
```
